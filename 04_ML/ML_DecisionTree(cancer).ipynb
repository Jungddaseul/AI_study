```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import matplotlib
```

```python
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
X[0:5], y[:5]
```

```python
cancer.feature_names
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                        stratify=cancer.target, 
                                        test_size = 0.3,  # 테스트 비율 30%
                                        random_state=77)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier  

# 모델 생성
tree = DecisionTreeClassifier(max_depth=2, random_state=0)

# 모델 학습
tree.fit(X_train, y_train)  

# 모델 평가
print("학습용 세트 정확도 : {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도 : {:.3f}".format(tree.score(X_test, y_test)))
```

### 의사결정트리 시각화
- binary dicision tree(이진 의사 결정 트리)

```python
from sklearn.tree import export_graphviz
import graphviz

export_graphviz(tree,
                out_file="tree.dot",
                class_names = ['악성', '양성'],
                feature_names = cancer.feature_names,
                impurity = False,  # gini 계수
                filled = True)     # color
  
with open("tree.dot") as f:
  dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_feature_imp_cancer(model):
  n_features = cancer.data.shape[1]
  imp = model.feature_importances_
  plt.barh(range(n_features) , imp, align='center')
  plt.yticks(np.arange(n_features), cancer.feature_names)
  
  plt.xlabel("feature importance")
  plt.ylabel("feature")
  plt.ylim(-1, n_features)

plt.figure(figsize=(12,12))
plot_feature_imp_cancer(tree)
```
